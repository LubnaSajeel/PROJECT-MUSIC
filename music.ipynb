```python
df1=pd.get_dummies(df[['Title','Artist','Top Genre']],drop_first=True)
df1
```

```python
data = df[["Beats Per Minute (BPM)","Energy","Loudness (dB)", 
              "Liveness", "Valence", "Acousticness", 
              "Speechiness"]]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(df)

```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  model=KMeans(n_clusters=i,init='k-means++',random_state=42)
  model.fit(df3)
  wcss.append(model.inertia_)
wcss
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11),wcss)
```

```python
model=KMeans(n_clusters=5,init='k-means++',random_state=42)
y_model=model.fit_predict(data)
y_model
```

```python
df3['cluster']=y_model
df3
```

```python

x=df3.iloc[:,:-1]
y=df3.iloc[:,-1]
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=1)
x_train
```

```python
plt.hist(x_train)
plt.show()
```

```python
plt.plot(x_train,linestyle='dotted',color='b')
plt.show()
```

```python
plt.plot(x_test,linestyle='dashed',color='r')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
score=accuracy_score(y_test,y_pred)
score
```

```python
matrix=confusion_matrix(y_test,y_pred)
matrix

```

```python
report=classification_report(y_test,y_pred)
report
```
